```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as dates
```

```python
data = pd.read_csv('Data/vix.csv', index_col='Date', parse_dates = True)
```

```python
data.head()
```

```python
data.tail()
```

```python
data_2021 = data.loc['2021']
```

```python
data_2020 = data.loc['2020-01-01':'2020-04-13']
```

```python
data_2020.reset_index()['VIX']
```

```python
plt.figure(figsize=(12,6))
plt.plot(data_2020.reset_index()['VIX'], 'b-', label='VIX: 2020 - (~32 avg)', lw=3)
plt.plot(data_2021.reset_index()['VIX'], 'r--', label='VIX: 2021 - (~22 avg)',lw=3)
plt.title('VIX YTD: 2021 vs. 2020')
plt.legend()
```

```python
sum(data_2020['VIX'])/len(data_2020)
```

```python
sum(data_2021['VIX'])/len(data_2021)
```

```python
idx_2020 = data_2020.index
vix_2020 = data_2020['VIX']
idx_2021 = data_2021.index
vix_2021 = data_2021['VIX']
```

```python
fig, ax = plt.subplots(figsize=(12,6))
ax.plot_date(idx_2020[:-1], vix_2020[:-1],'-', c='r', label='VIX: 2020', lw=3)
ax.plot_date(idx_2020[:-1], vix_2021,'-', c='b', label='VIX: 2021', lw=3)

# Major Axis
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%B'))

# Minor Axis
ax.xaxis.set_minor_locator(dates.WeekdayLocator())
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

# Grids
ax.yaxis.grid(True)
ax.xaxis.grid(True)

fig.autofmt_xdate() # Auto fixes the overlap!
plt.tight_layout()
plt.title('VIX: 2020 vs. 2021')
plt.legend()
plt.show()
```
