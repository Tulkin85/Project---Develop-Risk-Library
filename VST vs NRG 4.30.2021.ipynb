```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
from matplotlib import style
style.use('ggplot')
```

```python
df = pd.read_csv('Data\\vst_vs_nrg.csv', parse_dates=True, index_col='Dates')
#df = df[['AES', 'IBE']]
df.head()
```

```python
fig = plt.figure(figsize=(12,10))
ax1 = fig.add_subplot(2, 1,1)
ax1.plot(df['VST'], label='VST Price')
ax1.set_title('VST Price')
ax1.legend(loc='best')


ax2 = fig.add_subplot(2, 1,2)
ax2.plot(df['NRG'], label='NRG Price', color='g')
ax2.set_title('NRG Price')
ax2.legend(loc='best')
```

```python
df['Ratio'] = df['VST'] / df['NRG']
```

```python
def zscore(series):
    return (series - series.mean()) / np.std(series)
```

```python
zscore(df['Ratio'].loc['2018':]).plot(figsize=(14,8), color='darkblue')
plt.axhline(zscore(df['Ratio']).mean())
plt.axhline(1.0, color='red', linestyle='--')
plt.axhline(-1.0, color='green', linestyle='--')
plt.title('VST vs. NRG Trading Spread')
plt.show();
```

```python
df['ZScore']= zscore(df['Ratio'])
```

```python
df['ZScore'].plot(figsize=(12,8))
```
